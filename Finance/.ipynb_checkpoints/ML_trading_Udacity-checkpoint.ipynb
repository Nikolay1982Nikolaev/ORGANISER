```python
def get_max_close(symbol):
    """return the maximum closing value for stock indicated by symbol"""
    df = data.get_data_yahoo(symbol, start, end)
    df.to_csv(f'data/{symbol}.csv')
    df = pd.read_csv(f"data/{symbol}.csv")  # read in data
    return df['Close'].max()                #compute and return max

def test_run():
    """function called by Test Run"""
    for symbol in ['AAPL', '^GSPC']:
        print("Max Close")
        print(symbol, get_max_close(symbol))

        
if __name__ == "__main__":
    test_run()
```

```python
def get_mean_volume(symbol):
    df = data.get_data_yahoo(symbol, start, end)
    df.to_csv(f'data/{symbol}.csv')
    df = pd.read_csv(f"data/{symbol}.csv")  # read in data
    return df['Close'].mean() 


def test_run():
    """function called by Test Run"""
    for symbol in ['AAPL', '^GSPC']:
        print("Mean Close")
        print(symbol, get_mean_volume(symbol))

        
if __name__ == "__main__":
    test_run()
```

```python
import os
# Build a dataframe in pandas
def symbol_to_path(symbol, base_dir='data'):
    """Return CSV file path given ticker symbol"""
    return os.path.join(base_dir, f"{str(symbol)}.csv")


def get_Data(symbols, dates):
    """Read stock data (adjusting close) for given symbols form CSV files"""
    df = pd.DataFrame(index=dates)
    if 'AAPL' not in symbols: #add AAPL for reference, is absent
        symbols.insert(0, 'AAPL')
    
        #read in more stocks
    
    for symbol in symbols:
        df = data.get_data_yahoo(symbol)
        #print(df)
        df.to_csv(f'data/{symbol}.csv')
        df_temp = pd.read_csv(f"data/{symbol}.csv", index_col='Date',parse_dates=True, usecols=['Date', 'Adj Close'],na_values=['nan'])
   
    #rename to prevent clash
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df = df.join(df_temp) #use default how='left'
        if symbol == 'SPY': #drop dates SPY did not trade
            df = df.dropna(subset=['SPY'])
    return df


def test_run():
    #define a data range
    dates = pd.date_range('2010-01-01', "2020-01-01")
    
    # choose stock symbols to read
    symbols = ['AAPL', "^RUT", "^IXIC"] # SPY will be adde in get_data()
    
    #get stock data
    df = get_Data(symbols, dates)
    print(df.ix["2020-01-01":"2018-01-31"])
    
    #slice by column (symbols)
    print(df['AAPL'])  # a single label select a single column
    print(df[['AAPL', "^RUT"]])
    
    # slice by row and column
    print(df.ix["2010-03-10":"2010-03-15", ['AAPL', "^RUT"]])
    
    
if __name__ == "__main__":
    test_run() 
```
