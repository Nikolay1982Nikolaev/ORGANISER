## Fundamental Concepts
### ML

### Types of ML
- SUpervised Learning
    - we train the program on previously know labelled data
    - after training, we expect it to make predictions on new data
    - ex: regression, classification
- Unsupervised Learning:
    - we leave the prgram to find patterns in data
    - ex: clustering analysis, dimensionality reduction
- Reinforcement Learning:
    - a form of supervised learning
    - the program learns continuously
    - ex: learning to play a game by observing other players, learning to drive a car

### Algorithms by Task
