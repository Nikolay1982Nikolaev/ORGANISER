- f = open('file_name', 'mode', 'buffer')

- Modes: 
    - w, r, a, 
    - w+: write and read
    - r+: read 
    - a+: append and read
    - x: new file

- f('close')

## Write to a file

```python
# open the file for writting
f = open('my_file.txt', 'w')
s = input('Enter text: ')
f.write(s)
f.close()
```

## Read from a file

```python
f = open('my_file.txt', 'r')
s = f.read()
print(s)
f.close()
```

## Write multiple string:

```python
f = open('my_file.txt', 'w')
print('Enter text (Type # when you are done)')

s = ''
while s != '#':
    s = input('Enter text: ')
    f.write(s + '\n')

f.close()
```

## Check if the file exists

```python
import os, sys

if os.path.isfile('my_file.txt'):
    f = open('my_file.txt', 'r')
    s = f.read()
    print(s)
    f.close()
    
else:
    print('File Does not exist')
    sys.exit()
   

s = f.read()
print(s)
f.close()
```
