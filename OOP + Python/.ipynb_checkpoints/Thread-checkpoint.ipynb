### Using function:

```python
from threading import *
```

```python
def display_numbers():
    i = 0
    while i < 10:
        print(i)
        i += 1

t = Thread(target= display_numbers)
t.start()
```

```python
# printing Thread Names
def display_numbers():
    i = 0
    print(current_thread().getName())
    while i < 10:
        
        print(i)
        i += 1

print(current_thread().getName())
t = Thread(target= display_numbers)
t.start()
```

### using superclass

```python
from threading import Thread

class MyThread(Thread):
    def run(self):
        i = 0
        print(current_thread().getName())
        while i < 10:
            print(i)
            i += 1

            
t = MyThread()
t.start()
```

## Using a Class

```python
from threading import *

class MyThread:
    
    def display(self):
        i = 0
        print(current_thread().getName())
        while i < 10:
            print(i)
            i += 1

            
obj = MyThread()
t = Thread(target= obj.display())
t.start()
```

## Multithreading in action:

```python
from threading import *
from time import sleep

class MyThread:
    
    def display(self):
        i = 0
        print(current_thread().getName())
        while i < 10:
            print(i)
            i += 1

            
obj = MyThread()
t = Thread(target= obj.display())
t.start()
t2 = Thread(target= obj.display())
t2.start()
t3 = Thread(target= obj.display())
t3.start()
```

## Using Sleep

```python
from threading import *
from time import sleep

class MyThread:
    
    def display(self):
        i = 0
        print(current_thread().getName())
        sleep(1)
        while i < 10:
            print(i)
            i += 1

            
obj = MyThread()
t = Thread(target= obj.display())
t.start()
t2 = Thread(target= obj.display())
t2.start()
```

## TicketBooking usecase:

```python
from threading import *

class BookMyBus:
    def buy(self):
        print('Confirming a seat')
        print('Processing the payment')
        print('Printing the Ticket')
      
    
obj = BookMyBus()
t1 = Thread(target= obj.buy())
t2 = Thread(target= obj.buy())
t3 = Thread(target= obj.buy())

t1.start()
t2.start()
t3.start()
```

## Thread Synchronization:

```python
from threading import *

class BookMyBus:
    def __init__(self, available_seat):
        self.available_seat = available_seat
        #self.l = Lock()
        self.l = Semaphore()
        
    def buy(self, seat_requested):
        self.l.acquire()
        print('Total seats available: ', self.available_seat)
        
        if (self.available_seat >= seat_requested):
            print('Confirming a seat')
            print('Processing the payment')
            print('Printing the Ticket')
            self.available_seat -= seat_requested
        else:
            print('Sorry, not seat available')
        self.l.release()
    
obj = BookMyBus(8)
t1 = Thread(target= obj.buy, args=(3,))
t2 = Thread(target= obj.buy, args=(3,))
t3 = Thread(target= obj.buy, args=(3,))

t1.start()
t2.start()
t3.start()
```

```python
from threading import *
from time import *
```

```python
class Producer:
    def __init__(self):
        self.products = []
        self.ordersplaced = False
        
    def produce(self):
        for i in range(1,5):
            self.products.append('Product' + str(i))
            sleep(1)
            print('Item Added')
        #self.ordersplaced = True
        self.c.notify()
        self.c.release()

class Consumer:
    def __init__(self, prod):
        self.prod = prod
        
    def consume(self):
        self.prod.c.acquire()
        self.prod.c.wait(timeout=0)
        self.prod.c.release()
        print('Orders Shipped', self.prod.products)
        
p = Producer()
c = Consumer(p)

t1 = Thread(target=p.produce)
t2 = Thread(target=c.consume)

t1.start()
t2.start()
```
