#### The Scientific Method Steps
1. Ask a question
2. Do a reserch
3. Dorm a hypothesis
4. Test the hypothesis with an experiment
    - experiment works > analyze the data
    - experiment doesn't work > fix experiment
5. Result aligh with hypothesis > OK
6. Result don't align with hypothesis > new question, new hypothesis
7. Communicate the result

## Applied ML Process
- this allows us to do our job faster and more reliably
1. Probelm def.: make sure problem is well defined and that you're solving the right problem
2. Data analysis: get familiar with the available data
3. Data preparation: get the data ready for modelling
4. ALgorithm evaluation: test and compare algorithms
5. Result imrovement: use results to create better models (e.g. fine-tuning, ensembles)
6. Result presentation: describe the problem and solving to non- specialists

## Types of ML Algorithms:
1. Supervised learning: 
    - we train the program on previously known (labelled) data
    - after train, we expect it to make prediction on new data
    - ex. regression, classification
2. Unsepervised learning:
    - we leave the program to find patterns in data
    - ex. clustering analysis, dimensionality reduction
3. Reinforcement learning:
    - a form of unsupervised learning
    - the program learns continuously
    - ex. learn to play a game by observing other players, learning to drive a car
    

### Algorithms by Task:
- Statistical algorithms
- Regression: prediction a continuous variable
- Classification: predicting class labels
- Clustering: finding compact groups of data points
- Dimensionality reduction: simplifying  the input data
- Recommendation: suggest items for users
- Optimization: minimize/maximize a target function
- Testing and improvement algorithms: helper algorithms to select, fine-tune and optimize other ML algorithms
- and more...

### Example: Preparing Data for Modelling
- Most models require two additional steps:
    - 1. COnvert categorical variables into indicato variables:
        - dataset = pd.get_dummies(dataset_
     - 2. Normalize values: if needed (e.g.scale all variables from 0 to 1 using min-max scaling, or use Z-scores)
- Perform other model-specific transformations:
    - e.g. your model may not work well with highly imbalanced data (when you look for anomalie)
- If possible, prepare several versions of the dataset
    - to see how transformation affects model performance
- **Describe and document the entire process!**
    - don't forget the rules for reproducible research
