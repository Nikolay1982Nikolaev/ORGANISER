```python
import re
```

```python
str = 'Take up one idea. One idea at a time'
result = re.search(r'o\w\w', str)
print(result.group())
```

```python
str = 'Take up one idea. one idea at a time'
result = re.findall(r'o\w\w', str)
print(result)
```

```python
str = 'Take up one idea. One idea at a time'
result = re.match(r'T\w\w', str)
print(result.group())
```

```python
str = 'Take 1 up one  23 idea. One idea 45  at a time'
```

```python
result = re.split(r'\d+', str)
print(result)
```

```python
result = re.sub(r'One', 'Two', str)
print(result)
```

```python
# Quantifiers
str = 'Take up one idea. One idea at a time'

result = re.findall(r'O\w{2}', str)
print(result)
```

```python
result = re.findall(r'O\w{1,2}', str)
print(result)
```

```python
# 22-11-2012
# re.findall(f'd{1,2}-d{1,2},d{4}, str)
str = 'Take up 01-03-2020 one idea. One idea 02-05-2020 at a time'
result = re.findall(r'\d{1,2}-\d{1,2}-\d{4}', str)
```

```python
print(result)
```

```python
# special characters
# \ ^ $ [...] [^...] (...)  (R|S)

result = re.search(r'^T\w+', str)
print(result.group())
```
