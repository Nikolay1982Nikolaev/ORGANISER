```python
# Combining Date and Time

from datetime import *
```

```python
d = date(2018, 7,21)
t = time(12,45)
dt = datetime.combine(d,t)
```

```python
# sorting date
from datetime import *
l = []

d1 = date(2016, 8, 12)
d2 = date(2017, 8, 12)
d3 = date(2018, 8, 12)
```

```python
l.append(d1)
l.append(d2)
l.append(d3)

l.sort()
```

```python
for d in l:
    print(d)
```

```python
# sleep
from datetime import date
import time 

l = []

d1 = date(2016, 8, 12)
d2 = date(2017, 8, 12)
d3 = date(2018, 8, 12)
l.append(d1)
l.append(d2)
l.append(d3)

l.sort()
time.sleep(3)

for d in l:
    print(d)
```

```python
# performance:
from datetime import date
import time 

start_time = time.perf_counter()
l = []

d1 = date(2016, 8, 12)
d2 = date(2017, 8, 12)
d3 = date(2018, 8, 12)
l.append(d1)
l.append(d2)
l.append(d3)

l.sort()
#time.sleep(3)

for d in l:
    print(d)
    
end_time = time.perf_counter()
print('execution - ', start_time - end_time)
```
