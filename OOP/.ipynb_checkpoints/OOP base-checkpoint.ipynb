```python
class Course:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
        
    def avarage(self):
        number_of_ratings = len(self.rating)
        avarage = sum(self.rating) / number_of_ratings
        print(f"Avarege = {avarage}")


        
p2 = Course('Java', [4,5,6])
print(p2.name)
print(p2.rating)
p2.avarage()
```

```python
class Programmer:
    def set_name(self, n):
        self.name = n
        
    def get_name(self):
        return self.name
    
    def set_salary(self, s):
        self.salary = s
        
    def get_salary(self):
        return self.salary
    
    def set_technologies(self, t):
        self.technologies = t
        
    def get_technologies(self):
        return self.technologies
```

```python
p1 = Programmer()
p1.set_name('John')
p1.set_salary('1000')
p1.set_technologies('Java')
```

```python
print(p1.get_name())
print(p1.get_salary())
print(p1.get_technologies())
```

# Static or Class Level fields

```python
class Student:
    major = 'CSE'
    
    def __init__(self, num, name):
        self.num = num
        self.name = name  
```

```python
p1 = Student(1, 'John')
p2 = Student(2, 'Niko')
```

```python
print(p1.major)
print(p1.name)
print(p2.major)
print(p2.name)
```

## Static Method

```python
class ObjectCounter:
    
    couter_object = 0
    
    def __init__(self):
        ObjectCounter.counter_object += 1
    
    @staticmethod
    def display_count(self):
        print(ObjectCounter.counter_object)
        
```

```python
o1 = ObjectCounter()
```

```python
class Car:
    def __init__(self, make, year):
        self.make = make
        self.year = year
        
    class Engine:
        def __init__(self, number):
            self.number = number
        def start(self):
            print('Engine Started')
        
```
